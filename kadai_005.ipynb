```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(2,3,tight_layout=True)

y0 = [1,2,-5,2]
axes[0,0].plot(y0)

x1 = np.linspace(0,10,100)
y1 = 2 + 2 *np.sin(2 * x1)
ax1 = plt.subplot(2,3,2)
ax1.plot(x1,y1,linewidth=2.0)
ax1.set(xlim=(0,10), xticks=np.arange(0,10), ylim=(-1,5), yticks=np.arange(-1, 6))

x2 = np.random.normal(15, 5, 2000)
ax2 = plt.subplot(2,3,3)
ax2 = plt.xlim(-10, 40)
ax2 = plt.hist(x2,bins=30)

np.random.seed(3)
x3 = 4 + np.random.normal(0,2,60)
y3 = 4 + np.random.normal(0, 2, len(x3))
ax3 = plt.subplot(2,3,4)
ax3 = plt.scatter(x3, y3)

plt.show()
```
